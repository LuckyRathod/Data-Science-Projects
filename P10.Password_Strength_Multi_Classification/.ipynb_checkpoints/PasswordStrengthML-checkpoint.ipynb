### Check password strength using ML Algorithms

- Weak -- Label 0 
- Medium -- Label 1 
- Strong -- Label 2 

```python
import pandas as pd
import numpy as np
data = pd.read_csv('data.csv',',',error_bad_lines=False) ## If Comma seperated values are not available then skip row
data.head()
```

### We will use NLP Techniques

```python
data.isnull().sum()
```

```python
data[data['password'].isnull()]
```

```python
data.dropna(inplace=True)
```

```python
password_tuple = np.array(data)
password_tuple
```

```python
import random
random.shuffle(password_tuple)  ## Shuffling Randomly for Robustness
password_tuple
```

```python
y = [label[1] for label in password_tuple]
y
```

```python
X = [label[0] for label in password_tuple]
X
```

```python
import seaborn as sns
sns.countplot(x='strength',data=data,palette='RdBu_r')
```

### Convert String to Vector 

```python
data.shape
```

```python
def word_divide_char(inputs):
    characters=[]
    for i in inputs:
        characters.append(i)
    return characters
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
## By Default words are considered in Tfidf Vectorizer tokenizer. Now in this case we have to work with Characters.
vectorizer = TfidfVectorizer(tokenizer=word_divide_char) ## Use Above Function instead of default ---Words
X = vectorizer.fit_transform(X)
```

#### 129 Features 

```python
X.shape
```

```python
vectorizer.vocabulary_
```

```python
X
```

```python
data.iloc[0,0]
```

### Now we will see , How above Password is converted into 129 Features

```python
feature_names = vectorizer.get_feature_names()
 
#get tfidf vector for first document
first_document_vector=X[0]
 
#print the scores
df = pd.DataFrame(first_document_vector.T.todense(), index=feature_names, columns=["tfidf"])
df.sort_values(by=["tfidf"],ascending=False)
```

```python
## Logistics Regression

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)  #splitting
```

#### ovr - One Vs Rest 
#### multinomial - Cross Entropy loss if multi_class is set to multinomial

```python
log_class=LogisticRegression(penalty='l2',multi_class='ovr',solver='liblinear') ### Logistic Regression for Multi class
log_class.fit(X_train,y_train)
```

```python
print(log_class.score(X_test,y_test))
```

```python
## Multinomial

clf = LogisticRegression(random_state=0, multi_class='multinomial', solver='newton-cg')
clf.fit(X_train, y_train) #training
print(clf.score(X_test, y_test))
```

```python
X_predict=np.array(["%@123abcd"])
X_predict=vectorizer.transform(X_predict)
y_pred=log_class.predict(X_predict)
print(y_pred)
```

### XG Boost Classifier

```python
import xgboost as xgb
xgb_classifier = xgb.XGBClassifier()
xgb_classifier.fit(X_train,y_train) 
```

```python
print(xgb_classifier.score(X_test, y_test))
```

```python
X_predict=np.array(["%@123abcd"])
X_predict=vectorizer.transform(X_predict)
y_pred=xgb_classifier.predict(X_predict)
print(y_pred)
```

### MultinomialNB

```python
from sklearn.naive_bayes import MultinomialNB
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train,y_train)
nb_classifier.score(X_test,y_test)
```

```python
X_predict=np.array(["%@123abcd"])
X_predict=vectorizer.transform(X_predict)
y_pred=nb_classifier.predict(X_predict)
print(y_pred)
```
